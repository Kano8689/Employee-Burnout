```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("burnout_dataset.csv")

df.head()
```

```python
df.columns = df.columns.str.replace(" ", "_")

df.columns
```

```python
df["Resource_Allocation"] = df["Resource_Allocation"].fillna(int(df["Resource_Allocation"].mean()))
df["Mental_Fatigue_Score"] = df["Mental_Fatigue_Score"].fillna(round(df["Mental_Fatigue_Score"].mean(), 1))

df.to_csv("burnout_dataset.csv")
```
